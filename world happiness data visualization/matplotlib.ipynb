```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import style
```

```python
data = pd.read_csv('2019.csv')
data
```

```python
plt.plot(data['Overall rank'],data['Social support'],color='g',label='lineOne',linewidth=2) 
plt.plot(data['Overall rank'],data['Freedom to make life choices'],color='b',label='lineTwo',linewidth=2)
plt.plot(data['Overall rank'],data['Generosity'],color='c',label='lineThree',linewidth=2)

plt.legend()
plt.title("Overall rank")
plt.xlabel("Overall rank(by country)")
plt.ylabel("features") 
```

```python
plt.bar(data['Overall rank'],data['Social support'],color='c',label='lineOne')

plt.legend()
plt.xlabel("Country")
plt.ylabel("Social support") 
```

```python
#stack plot

days =[1,2,3,4,5]
sleeping = [7,8,6,11,7]
eating = [2,3,4,3,2]
working = [7,8,7,2,2]
playing = [8,5,7,8,13]

plt.plot([],[],color='m',label=sleeping,linewidth=5)
plt.plot([],[],color='c',label=eating,linewidth=5)
plt.plot([],[],color='b',label=working,linewidth=5)
plt.plot([],[],color='g',label=playing,linewidth=5)

plt.stackplot(days,sleeping,eating,working,playing,colors=['m','c','b','g'])
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
#pie chart
slices=[5,10,15,30,60]
rank=["CEO","Head","Senior","mid-junior","junior"]
colors=['m','c','b','g','r']

plt.pie(slices,
        labels=rank,
        colors=colors,
        startangle=90,
        shadow=True,
        explode=(0,0,.1,0,0),
        autopct='%1.1f%%'
       )
plt.title("PIE PLOT")
plt.show()
```

```python
#Handling multiple plot
import numpy as np

def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)

t1 = np.arange(0.0,5.0,.1)
t2 = np.arange(0.0,5.0,0.02)

plt.subplot(211)
plt.plot(t1,f(t1),'bo',t2,f(t2))

plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2))
plt.show()
```
